## 3. ***Data Wrangling***

### Data Cleaning

````python
#columns having null_values
missing_values[missing_values>0]
````

````python
# It is better to drop the column 'company' because the number of missing values is extremely high.
hotel_df.drop(['company'],axis=1,inplace=True)
print(hotel_df.shape)
````

````python
# Filling/replacing null values in 'agent' and 'children' columns  with 0.
null_columns=['agent','children']
for col in null_columns:
  hotel_df[col].fillna(0,inplace=True)
print(hotel_df[['agent','children']].isnull().sum())
````

````python
# Filling/replacing null values in 'country' column with 'others'
hotel_df['country'].fillna('others',inplace=True)
print(hotel_df['country'].isnull().sum())
````

````python
#Let's check whether database having any other null values
## As we have seen, no columns has any null value
hotel_df.isnull().sum()
````

##**Change in datatype for required columns**

````python
# We have seen that children & agent column as datatype as float whereas it contains only int value, lets change datatype as 'int64'
hotel_df[['children', 'agent']] = hotel_df[['children', 'agent']].astype('int64')
````

 ***Adding new columns as per need ***

````python
#total stay in nights
hotel_df['total_stay_in_nights'] = hotel_df ['stays_in_week_nights'] + hotel_df ['stays_in_weekend_nights']
hotel_df['total_stay_in_nights']
````

````python
#creating new column 'revenue':
#revenue generated = total_stay_in_nights  * adr
# where, adr : Average Daily Rate, as defined by dividing the sum of all lodging transactions by the total number of staying nights
hotel_df['revenue'] = hotel_df['total_stay_in_nights'] *hotel_df['adr']
hotel_df['revenue']
````

````python
#Hotel wise total revenue:
hotel_df.groupby('hotel')['revenue'].sum()
````

````python
# adding  column  'total_guest' coming for each booking:
hotel_df['total_guest'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies']
hotel_df['total_guest'].sum()
````

````python
# for understanding, from col 'is_canceled': we will replace the value from (0,1) to not_canceled, is canceled.
hotel_df['is_canceled'] = hotel_df['is_canceled'].replace([0,1], ['not canceled', 'is canceled'])
hotel_df['is_canceled']
````

````python
#Same for 'is_repeated_guest' col
hotel_df['is_repeated_guest'] = hotel_df['is_repeated_guest'].replace([0,1], ['not repeated', 'repeated'])
hotel_df['is_repeated_guest']
````

What all manipulations have you done and insights you found?

### Data Manipulations Overview

In preparation for our analysis, several key manipulations were made to the dataset:

#### **Addition of Columns**
- **Total Guests:** To analyze guest volume and revenue, we introduced a `Total Guests` column. This column is calculated by summing the number of adults, children, and babies.
- **Revenue:** We created a `Revenue` column to assess the profitability and growth of each hotel. This is calculated by multiplying the `adr` (average daily rate) by the `Total Guests`.

#### **Removal of Columns**
- **Company:** We observed that the `Company` column contains mostly null values, rendering it irrelevant for our analysis. Consequently, we removed this column.

#### **Value Replacement in Columns**
- **Booking Status and Guest Type:** The `is_canceled`, `is_not_canceled`, and `is_repeated_guest` columns originally contained binary values (0,1) to indicate booking status and guest type. For clarity in visualization:
  - We replaced the values in `is_canceled` with 'Canceled' and 'Not canceled'.
  - Similarly, we replaced the values in `is_repeated_guest` with 'Repeated' and 'Not repeated'.

#### **Data Type Adjustments**
- **Agent & Children:** The `Agent` and `Children` columns contained float values, which were inappropriate since they represent agent IDs and guest counts. We converted these columns from float to integer types for accuracy.

#### **Handling Missing Values and Duplicate Entries**
- **Null Values:** Prior to visualization, we performed data wrangling. We identified and addressed columns with null values:
  - Columns with a significant number of null values were dropped, such as the `Company` column.
  - For columns with minimal null values, we used the `.fillna()` method to impute these missing values as required.
- **Duplicate Entries:** We also checked for and removed any duplicate rows using the `.drop_duplicates()` method to ensure the data's integrity.

---

In this way, we have removed unnecessary data & make our data clean and ready to analyse.

#### Chart - 4

````python
# Chart - 4 visualization code
# Get the distribution channel counts
distribution_counts = hotel_df['distribution_channel'].value_counts()

# Create a DataFrame with counts and calculate percentages
distribution_channel_df = pd.DataFrame({
    'distribution_channel': distribution_counts.index,
    'count': distribution_counts.values
})

# Calculate the percentage of each distribution channel
distribution_channel_df['percentage'] = round(distribution_channel_df['count'] * 100 / hotel_df.shape[0], 1)

# Create a list of percentages and labels for the pie chart
sizes = distribution_channel_df['percentage'].tolist()
labels = distribution_channel_df['distribution_channel'].tolist()

# Plot the pie chart
plt.figure(figsize=(15, 8))
plt.pie(
    distribution_channel_df['count'],
    explode=[0.05] * len(distribution_channel_df),
    shadow=False,
    startangle=90,
    colors=plt.cm.Paired.colors,
    labels=None,
    autopct='%1.1f%%',
    textprops={'fontsize': 10}
)

# Set up legends with the distribution channel names and percentages
labels_with_percentage = [f'{label}, {size}%' for label, size in zip(labels, sizes)]
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels_with_percentage)

# Set the title of the pie chart
plt.title('Mostly Used Distribution Channels for Hotel Bookings')

# Display the plot
plt.show()
````

#### Chart - 6

**ADR across different month**

````python
# Group by month and hotel, and calculate the mean ADR
bookings_by_months_df = (hotel_df.groupby(['arrival_date_month', 'hotel'])['adr']
                         .mean()
                         .reset_index())

# Define the order of months
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
                 'September', 'October', 'November', 'December']

# Ensure 'arrival_date_month' is treated as a categorical variable with the specified order
bookings_by_months_df['arrival_date_month'] = pd.Categorical(
    bookings_by_months_df['arrival_date_month'],
    categories=months_order,
    ordered=True
)

# Sort the DataFrame by 'arrival_date_month'
bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')

# Display the DataFrame
print(bookings_by_months_df)
````
